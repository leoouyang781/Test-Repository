```python
## Here's a more complex graph

import datetime
import matplotlib.pyplot as plt
import yfinance as yf

from matplotlib.dates import MonthLocator,DateFormatter

ticker='AAPL'
begdate= datetime.date( 2012, 1, 2 )

enddate = datetime.date( 2013, 12, 5)
months = MonthLocator(range(1,13), bymonthday=1, interval=3) # every 3rd month
monthsFmt = DateFormatter("%b '%Y")
x = yf.download(ticker, begdate, enddate)

x
```

```python
if len(x) == 0:
    print ('Found no quotes')
    raise SystemExit
dates = x.index
dates

closes = x.Close

fig, ax = plt.subplots(figsize = (10,6))
ax.plot_date(dates, closes, '-')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(monthsFmt)
ax.set_title("APPL stock from 2012-2013")
# ax.xaxis.set_minor_locator(mondays)
ax.grid(True)
fig.autofmt_xdate()
ax.autoscale_view()
```

### Intro to Statsmodels

statsmodels is a powerful Python package for many types of statistical analysis.

In statistics, ordinary least square (OLS) regression is a method for
estimating the unknown parameters in a linear regression model. It minimizes the sum
of squared vertical distances between the observed values and the values predicted
by the linear approximation. The OLS method is used extensively in finance. 

An example is shown below:

```python
"""
  Name     : c2_05_example_statsmodel_OLS.py
  Book     : Python for Finance (2nd ed.)
  Publisher: Packt Publishing Ltd. 
  Author   : Yuxing Yan
  Date     : 6/6/2017
  email    : yany@canisius.edu
             paulyxy@hotmail.com
"""

import numpy as np
import statsmodels.api as sm

y=[1,2,3,4,2,3,4]
x=range(1,8)
x=sm.add_constant(x)
results=sm.OLS(y,x).fit()
print(results.params)
```

```python
import scipy as sp
ret=np.array([0.05,0.11,-0.03])
pow(np.prod(ret+1),1/len(ret))-1
```

```python
def means (ret):
    return np.mean(ret), pow(np.prod(ret+1),1./len(ret))-1 

test = np.array([0.1,0.05,-0.02])

arith = means(test)[0]
geo = means(test)[1]
print('Arithmetic mean: {:.4} \nGeometric mean: {:.4}'.format(arith, geo))
```
